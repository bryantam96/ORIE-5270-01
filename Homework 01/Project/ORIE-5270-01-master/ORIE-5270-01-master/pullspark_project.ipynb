```python
import re, os, sys
import pandas as pd
from pyspark import SparkConf, SparkContext
```

```python
files = ['taqmaster20141201','taqmaster20141202','taqmaster20141203','taqmaster20141204','taqmaster20141205',\
         'taqmaster20141208','taqmaster20141209','taqmaster20141210','taqmaster20141211','taqmaster20141212',\
         'taqmaster20141215','taqmaster20141216','taqmaster20141217','taqmaster20141218','taqmaster20141219',\
         'taqmaster20141222','taqmaster20141223','taqmaster20141224','taqmaster20141226','taqmaster20141229',\
         'taqmaster20141230','taqmaster20141231']
lines = files
```

```python
memory = '20g'
pyspark_submit_args = ' --driver-memory ' + memory + ' pyspark-shell'
os.environ["PYSPARK_SUBMIT_ARGS"] = pyspark_submit_args
conf = SparkConf()
sc = SparkContext(conf=conf)

for i in range(len(lines)): lines[i] = sc.textFile(lines[i])
csval = lines
for i in range(len(csval)): csval[i] = lines[i].map(lambda x: re.split(';', x))

def splitter(elem1):
    ret = []
    ret.append(elem1[0][1:14])
    ret.append(elem1[0][14:26].rstrip())
    ret.append(elem1[1].rstrip())
    ret.append(elem1[2].rstrip())
    #ret.append(elem1[3][0:1])
    #ret.append(elem1[3][1:5])
    #ret.append(elem1[3][5:9])
    #ret.append(elem1[3][9:11])
    #ret.append(elem1[3][11:13].rstrip())
    #ret.append(elem1[4][0:4])
    #ret.append(elem1[4][4:5])
    #ret.append(elem1[4][5:13])
    #ret.append(elem1[4][13:21])
    #ret.append(elem1[4][21:29])
    #ret.append(elem1[4][29:37])
    #ret.append(elem1[4][37:38])
    #ret.append(elem1[4][38:39])
    #ret.append(elem1[4][39:42].rstrip())
    ret.append(elem1[5].rstrip())
    #ret.append(elem1[6][0:10].rstrip())
    #ret.append(elem1[6][10:11].rstrip())
    #ret.append(elem1[6][11:12].rstrip())
    #ret.append(elem1[7][0:1].rstrip())
    ret.append(elem1[7][1:5].rstrip())
    ret.append(elem1[7][5:15])
    #ret.append(elem1[7][15:19])
    #ret.append(elem1[7][19:20].rstrip())
    #ret.append(elem1[7][20:21].rstrip())
    #ret.append(elem1[7][21:29].rstrip())
    ret.append(elem1[7][29:30].rstrip())
    ret.append(elem1[7][30:31].rstrip())
    ret.append(elem1[7][31:32].rstrip())
    ret.append(elem1[8][0:1].rstrip())
    ret.append(elem1[8][1:2].rstrip())
    ret.append(elem1[8][2:3].rstrip())
    ret.append(elem1[8][3:4].rstrip())
    ret.append(elem1[8][4:5].rstrip())

    return ret

csvfn = csval
for i in range(len(csvfn)): csvfn[i] = csval[i].map(splitter)
data1 = csvfn
for i in range(len(data1)): data1[i] = pd.DataFrame(csvfn[i].collect())

for i in range(len(data1)):
    data1[i].columns = ['CUSIP','Symbol OTC','Symbol Compressed','Security Description',\
                     'Source - Listed Exchange','NYSE Industry Code','Shares Outstanding','NYSE Trade Today',\
                     'AMEX Trade Today','Boston Trade Today','NYSE Arca Trade Today','Philadelphia Trade Today',\
                     'NASD Trade Today','OTC Trade Today','CBOE Trade Today']
    data1[i] = data1[i].drop([0]).reset_index(drop=True)

#data1.columns = ['CUSIP','Symbol OTC','Symbol Compressed','Security Description','Unit of Trading',\
#                 'Specialist Clearing Agent','Specialist Clearing Number','Specialist Panel','Reason Code',\
#                 'Status Indicator','Status Date','Date Admitted Trading','Date Traded - Reg Way',\
#                 'Date Traded - Settlement Date','CCS Indicator','Fail Clearance Eligible','Reason Code',\
#                 'Source - Listed Exchange','Symbol','Source Info','Local Ind.','ITS Issue','NYSE Industry Code',\
#                 'Shares Outstanding','Unit of Trade','Denominator','Issue Type','Listing Type','NYSE Trade Today',\
#                 'AMEX Trade Today','Boston Trade Today','NYSE Arca Trade Today','Philadelphia Trade Today',\
#                 'NASD Trade Today','OTC Trade Today','CBOE Trade Today']

sc.stop()
```
